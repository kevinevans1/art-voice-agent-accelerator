```python
from typing import Any, Optional, List, Literal, Dict
import logging 
import uuid

logger = logging.getLogger(__name__)
logging.basicConfig(level=logging.INFO)
```

```python
fake_db = {}

class SchoolManagemntsystem: 
    """A class representing the school management system."""
    def __init__(self):
        self.records_upserted = 0
        
    def add_allumn_to_db(self, 
                         alumn: 'Alumn') -> None:
        try: 
            fake_db.setdefault(alumn.id, alumn)
        except AttributeError as e:
            logger.error(f"AttributeError: {e}")
        self.records_upserted += 1
        return
        
    def add_proffesor_to_db(self, 
                            proffesor: 'Proffesor') -> None:
        try: 
            fake_db.setdefault(proffesor.id, proffesor)
        except AttributeError as e:
            logger.error(f"AttributeError: {e}")
        self.records_upserted += 1
        return
    
    def retrieve_all_records(self) -> Dict[str, Any]:
        return fake_db
    
    def retrieve_record_by_name(self, 
                            name: str) -> Optional[Any]:
        for record in fake_db.values():
            if record.name == name:
                return record
        return None

class Person:
    """definition of a person in the school system"""
    def __init__(self, id: int, 
                 name: str,
                 age:str,
                 email:str):
        self.id = id 
        self.name = name
        self.age = age
        self.email = email
        
    def __repr__(self) -> str:
        return (f"Person(id={self.id}, "
                f"name={self.name}, "
                f"age={self.age}, "
                f"email={self.email})")

class Proffesor(Person): 
    """A class represing a proffesor with a name, 
    subjects taught, and optional years of experience."""
    def super().__init__(self, name: str, 
                 subjects: List[str], 
                 years_of_experience: Optional[int] = None):
        self.id = str(uuid.uuid4())
        self.name = name
        self.subjects = subjects
        self.years_of_experience = years_of_experience
        
    def __repr__(self) -> str:
        return (f"Proffesor(name={self.name}, "
                f"subjects={self.subjects}, "
                f"years_of_experience={self.years_of_experience})")
        
    def fetch_details(self) -> Dict[str, Any]:
        """Fetch detailed information about the proffesor."""
        details = {
            "name": self.name,
            "subjects": self.subjects,
            "years_of_experience": self.years_of_experience
        }
        logger.info(f"Fetched details for proffesor {self.name}")
        return details
    
    def upsert_experience(self, years: int) -> None:
        """Update or set the years of experience."""
        self.years_of_experience = years
        logger.info(f"Updated years of experience for {self.name} to {years}")
        
    def upsert_subjects(self, subjects: List[str]) -> None:
        """Update or set the subjects taught."""
        self.subjects = subjects
        logger.info(f"Updated subjects for {self.name} to {subjects}")
        


class Alumn(Person): 
    """A class represing an Alumn with a name, 
    subjects taught, and optional years of experience."""
    def __init__(self, name: str, 
                 subjects: List[str], 
                 profile: Optional[Literal["superior","intermediate","beginner","N/A" ]] = "N/A",
                 expected_graduation: Optional[int] = None, 
                 enrolled_courses: Optional[List[str]] = None):
        self.id = str(uuid.uuid4())
        self.name = name
        self.subjects = subjects
        self.profile = profile
        self.expected_graduation = expected_graduation
        self.enrolled_courses = enrolled_courses
        
    def __repr__(self) -> str:
        return (f"Alumn(name={self.name}, "
                f"subjects={self.subjects}, "
                f"profile={self.profile}, "
                f"expected_graduation={self.expected_graduation}, "
                f"enrolled_courses={self.enrolled_courses})")
        
    def fetch_details(self) -> Dict[str, Any]:
        """Fetch detailed information about the proffesor."""
        details = {
            "name": self.name,
            "subjects": self.subjects,
            "years_of_experience": self.years_of_experience
        }
        logger.info(f"Fetched details for proffesor {self.name}")
        return details
    
    def upsert_experience(self, years: int) -> None:
        """Update or set the years of experience."""
        self.years_of_experience = years
        logger.info(f"Updated years of experience for {self.name} to {years}")
        
    def upsert_subjects(self, subjects: List[str]) -> None:
        """Update or set the subjects taught."""
        self.subjects = subjects
        logger.info(f"Updated subjects for {self.name} to {subjects}")
        


professor_pablo = Proffesor(
    name="Pablo",
    subjects=["AI platforms"],
    years_of_experience=5
)
alumn_john = Alumn(
    name="John",
    subjects=["Math", "Science"],
    profile="intermediate",
    expected_graduation=2025,
    enrolled_courses=["Math", "Science"]
)
school_system = SchoolManagemntsystem()
school_system.add_proffesor_to_db(professor_pablo)
school_system.add_allumn_to_db(alumn_john)
```
